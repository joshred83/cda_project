```python
# Loaded variable 'df' from URI: /home/red/cda_project/data/bls_all_data_M_2022_metro.csv
import pandas as pd
import seaborn as sns

df = pd.read_csv(r"/home/red/cda_project/data/bls_all_data_M_2022_metro.csv")

# Drop column: 'I_GROUP' 'NAICS', 'NAICS_TITLE', 'OWN_CODE'
# Drop column: 'OCC_CODE' 'O_GROUP', 'OCC_TITLE' 'ANNUAL',
# Drop column: 'HOURLY'  'PCT_RPT', 'PCT_TOTAL' 'PCT_RPT',
# Drop column: 'PCT_TOTAL' 'AREA_TYPE' 'AREA_TYPE'


df = df.drop(
    columns=[
        "I_GROUP",
        "NAICS",
        "NAICS_TITLE",
        "OWN_CODE",
        "OCC_CODE",
        "O_GROUP",
        "OCC_TITLE",
        "ANNUAL",
        "HOURLY",
        "PCT_RPT",
        "PCT_TOTAL",
        "AREA_TYPE",
    ]
)

# Change column type to float64 for column: 'TOT_EMP'
numeric_columns = [
    "TOT_EMP",
    "EMP_PRSE",
    "JOBS_1000",
    "LOC_QUOTIENT",
    "H_MEAN",
    "A_MEAN",
    "MEAN_PRSE",
    "H_PCT10",
    "H_PCT25",
    "H_MEDIAN",
    "H_PCT75",
    "H_PCT90",
    "A_PCT10",
    "A_PCT25",
    "A_MEDIAN",
    "A_PCT75",
    "A_PCT90",
]
df[numeric_columns] = (
    df[numeric_columns].replace("[(**)|*|#]", None, regex=True).astype("float64")
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df[numeric_columns].corr()
fig, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", ax=ax)
```
