```python
df = get_data()

# rf = RandomForestRegressor(
#     # n_estimators= 16,
    
#     # criterion = "squared_error",
#     # max_depth = 4,
#     # min_samples_split= 4,
#     # min_samples_leaf= 4,
#     # # min_weight_fraction_leaf
#     # max_features= .5,
#     # max_leaf_nodes = 8,
#     # # min_impurity_decrease= Float = 0,
#     # bootstrap = True,
#     # oob_score = True,
#     # n_jobs=-1,
#     # random_state=42,
#     # # verbose=1,
#     # # warm_start=True,
#     # # ccp_alpha = 0,
#     # # max_samples = .9
# )


model = custom_pipe(df, BayesianRidge
)


# Initialize KFold for cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

outputs =[]
oob_scores = []
print(time())
for train_idx, test_idx in kf.split(df):
    print(model["impute"].get_params())
    model.fit(df.loc[train_idx, :])
    print(time())
    oob_scores.append(model["impute"].estimator.oob_score_)
    outputs.append(model.transform(df.loc[test_idx, :]))
```

```python
x_in = df.loc[test_idx, :]
y_out = inverse_prep(model, outputs[-1])
```
